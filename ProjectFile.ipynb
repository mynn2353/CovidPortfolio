```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# filling missing values 

```python
data = pd.DataFrame(data.continent.fillna(0))
data
```

```python
data= pd.DataFrame(data.population.fillna(0))
data
```

# which continent had most cases

```python
most_cases = data.groupby('continent').agg({'total_cases':'sum'}).sort_values('total_cases',ascending = False)
most_cases
```

```python
sns.barplot(x='continent', y='total_cases',data = most_cases.reset_index())
plt.xticks(rotation ='vertical')
```

```python
data
```

# which continent has most population

```python
pop = pd.DataFrame(data.groupby('continent').sum()['population'])
pop.sort_values(by=['population'], inplace = True, ascending = False)
pop
```

```python
sns.barplot(x='continent',y='population', data = pop.reset_index())
plt.xticks(rotation = 'vertical')
```

# comparision between cases and population

```python
sns.barplot(x='continent', y='total_cases',data = most_cases.reset_index())
plt.xticks(rotation ='vertical')
```

```python
sns.barplot(x='continent',y='population', data = pop.reset_index())
plt.xticks(rotation = 'vertical')
```
